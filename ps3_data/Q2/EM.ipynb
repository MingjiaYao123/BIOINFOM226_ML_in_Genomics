```python
q_list=[]
for iteration in range(5):
    q=0
    for k in range(2):
        for i in range(1000):
            q=q+r_em.iloc[i,k]*math.log(pi_em[k])
    for k in range(2):
        for i in range(1000):
            for j in range(5000):
                x=mix1_geno.iloc[i,j]
                fr=freq_em.iloc[j,k]
                prob=(fr**x)*((1-fr)**(1-x))
                q=q+r_em.iloc[i,k]*math.log(prob)
    q_list.append(q)
    print(q)
    for i in range(1000):
        l1=math.log(pi_em[0])
        l2=math.log(pi_em[1])
        for j in range(5000):
            x=mix1_geno.iloc[i,j]
            fr1=freq_em.iloc[j,0]
            fr2=freq_em.iloc[j,1]
            l1=l1+ math.log((fr1**x)*((1-fr1)**(1-x)))
            l2=l2+ math.log((fr2**x)*((1-fr2)**(1-x)))
        l_star=max(l1,l2)
        pr1=math.exp(l1-l_star)
        pr2=math.exp(l2-l_star)
        pr_sum=pr1+pr2
        r_em.iloc[i,0]=pr1/pr_sum
        r_em.iloc[i,1]=pr2/pr_sum
    r_sum_1=sum(r_em['k_1'])
    r_sum_2=sum(r_em['k_2'])
    pi_em[0]=r_sum_1/(r_sum_1+r_sum_2)
    pi_em[1]=r_sum_2/(r_sum_1+r_sum_2)
    for j in range(5000):
        f_j1=0
        f_j2=0
        for i in range(1000):
            f_j1=f_j1+r_em.iloc[i,0]*mix1_geno.iloc[i,j]
            f_j2=f_j2+r_em.iloc[i,1]*mix1_geno.iloc[i,j]
        f_j1=f_j1/r_sum_1
        f_j2=f_j2/r_sum_2
        freq_em.iloc[j,0]=f_j1
        freq_em.iloc[j,1]=f_j2
    print(iteration)
```
